```python
import fuzzywuzzy
import rapidfuzz
import logging
import time
```

```python
class Timer:
    def __enter__(self):
        self.start = time.time()
        return self  # возвращаем self, чтобы получить доступ к elapsed_time

    def __exit__(self, exc_type, exc_val, exc_tb):
        self.end = time.time()
        self.elapsed = round(self.end - self.start,4)
        #print(f"Cell execution time: {self.elapsed:.4f} seconds")
```

```python
import fuzzywuzzy.process


logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

class BaseMatcher:
    def __init__(self, df, threshold=80):
        self.df = df
        self.threshold = threshold

    def find_best_match(self, ocr_name):
        raise NotImplementedError("This method should be overridden by subclasses")

class FuzzyWuzzySimpleMatcher(BaseMatcher):
    def __init__(self, df, threshold=80, **kwargs):
        self.df = df
        self.threshold = threshold
        self.kwargs = kwargs

    def find_best_match(self, ocr_name):
        choices = self.df['description'].tolist()
        result = fuzzywuzzy.process.extractOne(ocr_name, choices,  score_cutoff = self.threshold, **self.kwargs)
        if result is None:
            logging.warning(f"No match found for: {ocr_name}")
            return None

        match, score = result
        idx = choices.index(match)

        #logging.info(f"Match: {match}, Score: {score}")
        return self.df.iloc[idx],score

class FuzzyRapidSimpleMatcher(BaseMatcher):
    def __init__(self, df, threshold=80, **kwargs):
        self.df = df
        self.threshold = threshold
        self.kwargs = kwargs
    def find_best_match(self, ocr_name):
        choices = self.df['description'].tolist()
        result = rapidfuzz.process.extractOne(ocr_name, choices, score_cutoff=self.threshold, **self.kwargs)

        if result is None:
            logging.warning(f"No match found for: {ocr_name}")
            return None

        match, score, idx = result
        logging.info(f"Match: {match}, Score: {score}")
        return self.df.iloc[idx],score


class FuzzyRapidByWordsMatcher(BaseMatcher):
    def __init__(self, df, threshold=80, **kwargs):
        self.df = df
        self.threshold = threshold
        self.kwargs = kwargs
    
    def find_best_match(self, ocr_name):
        words = ocr_name.split()
        best_match = None
        best_score = 0
        for index, row in self.df.iterrows():
            name = row['description']
            score = rapidfuzz.fuzz.token_set_ratio(ocr_name, name)
            if score > best_score and score >= self.threshold:
                best_score = score
                best_match = row

        logging.info(f"Best match for {ocr_name}: {best_match['description']} with score {best_score}")
        if best_score >= self.threshold:
            return best_match
        return None

```

```python
def evaluate_products(x_test, df_y_test,  matcher):
    matcher 
    matches = []
    
    for index, expected_row in df_y_test.iterrows():
        print(expected_row,'---')
        ocr_name = x_test[index]
        print(ocr_name)
        with Timer() as t:
            match_result = matcher.find_best_match(ocr_name)
        
        if match_result is not None:
            match, score = match_result
            print(match, score)
            matched = False
            print(match["sku"])
            if str(match["sku"]) == str(expected_row["sku"]) :
                matched = True
                logging.info(f"Match found for {ocr_name}: {match['description']} with Article {match['sku']}")
            else:
                logging.warning(f"Match found for {ocr_name} but article does not match: {match['sku']} vs {expected_row['sku']}")

            matches.append({
                'Test Name': ocr_name,
                'Matched Name': match['description'],
                'sku': match['sku'],
                "expected sku": expected_row['sku'],
                "matched": matched,
                "score": score,
                "elapsed_time": t.elapsed
            })
          
        else:
            matched = False
            matches.append({
                'Test Name': ocr_name,
                'Matched Name': None,
                'sku': None,
                "expected sku": expected_row['sku'],
                "matched": matched,
                "score": None,
                "elapsed_time": t.elapsed
            })
            logging.warning(f"No match found for {ocr_name}")

    return pd.DataFrame(matches)
```

```python
def mass_test(x_test,  matcher):
    matcher 
    matches = []
    
    for index, ocr_text in enumerate(x_test):
        
        
        print(ocr_text)
        with Timer() as t:
            match_result = matcher.find_best_match(ocr_text)
        
        if match_result is not None:
            match, score = match_result
            print(match, score)
            
            input_description = ocr_text
            result_sku = match['sku']
            result_description = match['description']
            result_score = score

            
          
        else:
            input_description = ocr_text
            result_sku = None
            result_description = None
            result_score = None
            
        matches.append({
            'input_description': input_description,
            'result_description': result_description,
            'result_sku': result_sku,
            "result_score": result_score,
            "elapsed_time": t.elapsed
        })
    return pd.DataFrame(matches)
```

# Тестируем через api

```python
import requests
import json

# Конфигурация
BASE_URL = "http://matching-api:3000"  # Замените на ваш URL
ENDPOINT = "/api/matching"

def test_matching_service(query, threshold=80):
    """
    Отправляет запрос к сервису matching
    
    Args:
        query (str): Поисковый запрос
        threshold (int): Порог совпадения (по умолчанию 80)
    
    Returns:
        dict: Ответ от сервера
    """
    url = f"{BASE_URL}{ENDPOINT}"
    
    payload = {
        "query": query,
        "threshold": threshold
    }
    
    headers = {
        "Content-Type": "application/json"
    }
    
    try:
        response = requests.post(url, json=payload, headers=headers)
        
        # Проверяем статус ответа
        if response.status_code == 200:
            return {
                "success": True,
                "data": response.json()
            }
        else:
            return {
                "success": False,
                "status_code": response.status_code,
                "error": response.json() if response.content else "Нет содержимого"
            }
            
    except requests.exceptions.RequestException as e:
        return {
            "success": False,
            "error": f"Ошибка соединения: {str(e)}"
        }
```

```python
test_matching_service("Капуста белокочанная 1кг.", 20)
```

```python
# Создать отдельрный матчер для rest api
class RestApiMatcher(BaseMatcher):
    def __init__(self, base_url, endpoint, threshold=80):
        self.base_url = base_url
        self.endpoint = endpoint
        self.threshold = threshold

    def find_best_match(self, ocr_name):
        response = test_matching_service(ocr_name, self.threshold)
        
        if not response['success']:
            logging.error(f"Error in API response: {response['error']}")
            return None
        
        data = response['data']
        
        if not data:
            logging.warning(f"No match found for: {ocr_name}")
            return None
        
        match = data.get("result")  # Предполагаем, что первый элемент - лучший матч
        score = data.get('score')
        
        logging.info(f"Match: {match['description']}, Score: {score}")
        return match, score
```
